```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
x = np.array([[2,1],[6.5,2.4],[4,7],[3,5.5],[3,3.7]])
```

```python
print(x)
```

```python
Z = linkage(x)
```

```python
print(Z)
```

```python
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(Z)
plt.show()
```
